Custom training for word2vec


```python
import pandas as pd
import numpy as np
# for creating data frames and convert data into arrays
import os
import gensim
import nltk
```

```python
nltk.download("all")  # all required files
```

```python
from nltk import sent_tokenize # sent_tokenize--> Convert data into sentences
from gensim.utils import simple_preprocess
```

```python
story=[]  #creating a story
```

```python
for file in os.listdir("/content/data"):
  print(file)  # files in that directory
```

```python
# Now read data from file
f=open("/content/data/001ssb.txt",encoding='unicode_escape')
```

```python
corpus=f.read() # reading data
# data we call it as corpous
```

```python
# Convert the data into tokenizers/sentenses
raw_sent=sent_tokenize(corpus)
```

```python
# Simple preprocess--predefined method responsible for basic text processing
for i in raw_sent:
  print(simple_preprocess(i))
# each and every sentence converted into tokens
```

```python
# Convert all the data into tokens
for filename in os.listdir("/content/data"):
  file_path=os.path.join("/content/data",filename)
  with open(file_path,encoding='unicode_escape') as f:
    corpus=f.read()
  raw_sent=sent_tokenize(corpus)
  for sent in raw_sent:
    story.append(simple_preprocess(sent))
```

```python
story
```

```python
len(story) # No.of sentenses
```

```python
story[0]
```

```python
model=gensim.models.Word2Vec(window=10,min_count=5,vector_size=150)  #my model
#window= context window
# vector=dimension of a vector
```

```python
model.build_vocab(story)
# creating unique words from the tokens
```

```python
model.corpus_count
```

```python
model.epochs
```

```python
# 5 means 5times trained
```

```python
model.corpus_total_words
```

```python
model.min_count  # frequency of the word
```

```python
# model training
model.train(story,total_examples=model.corpus_count,epochs=5)  # training model for 5times
```
